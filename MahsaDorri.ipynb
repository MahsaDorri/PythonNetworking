```python
from tkinter import *
from tkinter import ttk
color="light green"

window = Tk()
window.geometry('400x330')
window.title("Centennial College")

# create a frame
frame = Frame(window, bg=color, bd=10)
frame.pack()

# add a title label
title = Label(frame, text="ICET Student Survey",font=("BOLD", 20),bg=color)
title.grid(column=0,row=0,columnspan=3)
blank_space = Label(frame,text="",bg=color)
blank_space.grid(column=0,row=1,columnspan=3)

# add labels for the 1st column
name_label = Label(frame, text="Full name:", font=15,bg=color)
name_label.grid(column=0,row=2)
residency_label = Label(frame, text="Residency:", font=15,bg=color)
residency_label.grid(column=0, row=3)
program_label = Label(frame, text="Program:", font=15,bg=color)
program_label.grid(column=0,row=4)
course_label = Label(frame, text="Course:", font=15,bg=color)
course_label.grid(column=0,row=5)

# entry that captures the username
name_textbox = Entry(frame, font=15)
name_textbox.grid(column=1,row=2,columnspan=2)

# add radio buttons
residence = StringVar()
residence.set("dom")
dom_rb = Radiobutton(frame, text="Domestic", variable=residence, value="dom",bg=color,font=12)
dom_rb.grid(column=1,row=3)
intl_rb = Radiobutton(frame, text="International", variable=residence, value="intl",bg=color,font=12)
intl_rb.grid(column=2,row=3)

# add combobox
program_combo = ttk.Combobox(frame,values=["AI","Gaming","Health","Software"],state="readonly",font=12)
program_combo.grid(column=1,row=4,columnspan=2)

# add checkbox
prog1 = BooleanVar()
prog2 = BooleanVar()
prog3 = BooleanVar()
prog1_checkbox = Checkbutton(frame, text="Programing I", variable=prog1,font=12,bg=color)
prog1_checkbox.grid(column=1, row=5,columnspan=2,sticky=W)
prog2_checkbox = Checkbutton(frame, text="Web Page Design", variable=prog2, font=12,bg=color)
prog2_checkbox.grid(column=1,row=6,columnspan=2,sticky=W)
prog3_checkbox = Checkbutton(frame, text="Software Engineering", variable=prog3, font=12,bg=color)
prog3_checkbox.grid(column=1,row=7,columnspan=2,sticky=W)

# reset function
def reset_clicked():
    name_textbox.delete(0,END)
    residence.set("dom")
    prog1.set(FALSE)
    prog2.set(FALSE)
    prog3.set(FALSE)
    program_combo.set('')

# Ok function
def ok_clicked():
    new_window = Toplevel(window)
    new_window.geometry('300x110')
    new_window.title('Information')
    name = Label(new_window, text=name_textbox.get())
    name.pack()
    program = Label(new_window,text=program_combo.get())
    program.pack()
    residency = Label(new_window, text=residence.get())
    residency.pack()
    my_program=[]
    if prog1.get():
        my_program.append("Programing")
    if prog2.get():
        my_program.append("Web Page Design")
    if prog3.get():
        my_program.append("Software Engineering")
    programs = Label(new_window,text=my_program)
    programs.pack()
    new_button = Button(new_window, text='Ok', command=new_window.destroy)
    new_button.pack()

# Reset
reset_button=Button(frame,text="Reset",font=15,width=8,command=reset_clicked)
reset_button.grid(column=0,row=8)

# ok
ok_button=Button(frame,text="Ok",font=15,width=8,command=ok_clicked)
ok_button.grid(column=1,row=8)

# Exit
exit_button=Button(frame,text="Exit",font=15,command=window.destroy,width=8)
exit_button.grid(column=2,row=8)

window.mainloop()
```
